```python
key = 'i4cUd4v7PHvpSJTNHJkCvjcQx2u%2FsHVqyzKdW490W7OUbcGzBT2rWaxtrrXj1XKWXyDoJK6L8KG%2B0CRp7kUVPQ%3D%3D'
number = '1'
date = '20170101'
url = 'http://apis.data.go.kr/B552895/LocalGovPriceInfoService/getItemPriceResearchSearch?serviceKey={0}&numOfRows=500&pageSize=500&pageNo={1}&startPage={2}&_returnType=xml%2Cjson&examin_de={3}'.format(key,number,number,date)
temp = urllib.request.urlopen(url).read().decode('utf-8')
temp_crawler = BeautifulSoup(temp, 'lxml')
```

```python
#원하는 기간 구간 만들기

date_range = pd.date_range(start = '12/12/2017', end = '1/11/2018')

str_date_range = []
for i in range(len(date_range)) :
    str_day = str(date_range[i])[0:4]+str(date_range[i])[5:7]+str(date_range[i])[8:10]
    str_date_range.append(str_day)
```

```python
# 원하는 기간 내 일자별 데이터 갯수 가져오기
num_of_data = []

for date in str_date_range:
        
        url = 'http://apis.data.go.kr/B552895/LocalGovPriceInfoService/getItemPriceResearchSearch?serviceKey={0}&numOfRows=500&pageSize=500&pageNo={1}&startPage={2}&_returnType=xml%2Cjson&examin_de={3}'.format(key,number,number,date)
        temp = urllib.request.urlopen(url).read().decode('utf-8')
        temp_crawler = BeautifulSoup(temp, 'lxml')
        
        totalcount = temp_crawler.find_all('totalcount')[0].get_text()
        num_of_data.append(totalcount)
```

```python
PriceList_sample = pd.DataFrame({'date': [i.text for i in temp_crawler.find_all('examin_de')],
                            'area_name': [i.text for i in temp_crawler.find_all('examin_area_nm')],
                            'area_code': [i.text for i in temp_crawler.find_all('examin_area_cd')],
                            'market_name': [i.text for i in temp_crawler.find_all('examin_mrkt_nm')],
                            'market_code': [i.text for i in temp_crawler.find_all('examin_mrkt_cd')],
                            'product_name': [i.text for i in temp_crawler.find_all('prdlst_nm')],
                            'product_code': [i.text for i in temp_crawler.find_all('prdlst_cd')],
                            'product_detail_name': [i.text for i in temp_crawler.find_all('prdlst_detail_nm')],
                            'product_detail_code': [i.text for i in temp_crawler.find_all('prdlst_detail_cd')],   
                            'distb_step_code': [i.text for i in temp_crawler.find_all('distb_step_se')],
                            'distb_step_name': [i.text for i in temp_crawler.find_all('distb_step')],
                            'grade_code': [i.text for i in temp_crawler.find_all('grad_cd')],   
                            'price_examined': [i.text for i in temp_crawler.find_all('examin_amt')],
                            'price_examined_before': [i.text for i in temp_crawler.find_all('bfrt_examin_amt')],     
                            'standard': [i.text for i in temp_crawler.find_all('stndrd')],     
                            'grade_name': [i.text for i in temp_crawler.find_all('grad')]})
```

```python
for i, date in enumerate(str_date_range):
    
    number = num_of_data[i]
    url = 'http://apis.data.go.kr/B552895/LocalGovPriceInfoService/getItemPriceResearchSearch?serviceKey={0}&numOfRows={1}&pageSize={2}&pageNo=1&startPage=1&_returnType=xml%2Cjson&examin_de={3}'.format(key,number,number,date)
    temp = urllib.request.urlopen(url).read().decode('utf-8')
    temp_crawler = BeautifulSoup(temp, 'lxml')
            
    PriceList_date = pd.DataFrame({'date': [i.text for i in temp_crawler.find_all('examin_de')],
                            'area_name': [i.text for i in temp_crawler.find_all('examin_area_nm')],
                            'area_code': [i.text for i in temp_crawler.find_all('examin_area_cd')],
                            'market_name': [i.text for i in temp_crawler.find_all('examin_mrkt_nm')],
                            'market_code': [i.text for i in temp_crawler.find_all('examin_mrkt_cd')],
                            'product_name': [i.text for i in temp_crawler.find_all('prdlst_nm')],
                            'product_code': [i.text for i in temp_crawler.find_all('prdlst_cd')],
                            'product_detail_name': [i.text for i in temp_crawler.find_all('prdlst_detail_nm')],
                            'product_detail_code': [i.text for i in temp_crawler.find_all('prdlst_detail_cd')],   
                            'distb_step_code': [i.text for i in temp_crawler.find_all('distb_step_se')],
                            'distb_step_name': [i.text for i in temp_crawler.find_all('distb_step')],
                            'grade_code': [i.text for i in temp_crawler.find_all('grad_cd')],   
                            'price_examined': [i.text for i in temp_crawler.find_all('examin_amt')],
                            'price_examined_before': [i.text for i in temp_crawler.find_all('bfrt_examin_amt')],     
                            'standard': [i.text for i in temp_crawler.find_all('stndrd')],     
                            'grade_name': [i.text for i in temp_crawler.find_all('grad')]})
    
    PriceList_date.to_csv('PriceList_{}.csv'.format(date))
    print('{}일 저장!'.format(date))
```
