```python
def process_price(df):
    '''같은 품목에 대하여 standard를 한 가지로 맞추어주는 함수입니다.\
    인풋으로 데이터 프레임을 받아 변환된 standard에 맞는 가격을 리턴합니다.
    
    
    
    Example:
        df['standard_processed'] = df.apply(process_standard, axis=1)
    '''
    if df.product_name == '콩' or df.product_name == '팥(적두)':
        if df.standard == '1KG': return df.price_examined / 2
        else: return df.price_examined
    elif df.product_name == '감자':
        if df.standard == '100G': return df.price_examined * 10
        else: return df.price_examined
    elif df.product_name == '포도':
        if df.standard == '2KG': return df.price_examined / 2
        else: return df.price_examined
    elif df.product_name == '계란':
        if df.standard == '10개': return df.price_examined * 3
        else: return df.price_examined
    elif df.product_name == '고등어':
        if df.standard == '2마리': return df.price_examined / 2
        else: return df.price_examined
    elif df.product_name == '깻잎':
        if df.standard == '200G': return df.price_examined / 2
        else: return df.price_examined
    else:
        return df.price_examined
```
